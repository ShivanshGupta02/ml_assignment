### Question 1
##### How many multiplications and additions do you need to perform a matrix multiplication between a (n, k) and (k, m) matrix? Explain.

###### Ans1 - For filling one position in a final (n, m) matrix, we have to do "k" multiplications and "k-1" additions. There are total m*n such operations need to be done in b/w of row of the first matrix and column of the second matrix. Hence for filling whole matrix, we have to do - 
              Total No. of Multiplications = m*n*k 
              Total No. of Additions = m*n*(k-1)

### Question 2
##### Write Python code to multiply the above two matrices. Solve using list of lists and then use numpy. Compare the timing of both solutions. Which one is faster? Why?

```python
n, m, k = 5, 6, 8

import timeit

# Using list of lists
mat1 = [[2 for i in range(0,k)] for j in range(0,n) ]
mat2 = [[3 for i in range(0,m)] for j in range(0,k) ]
mat = [[0 for i in range(0,m)] for j in range(0,n)]
start = timeit.default_timer()
for i in range(0,n):
    for j in range(0,m):
        for p in range(0,k):
            mat[i][j] += mat1[i][p]*mat2[p][j]
# print(mat)
end = timeit.default_timer()
print("Total time taken using list of lists : ",end-start)

# Using numpy
import numpy as np
mat1 = np.array([[2 for i in range(0,k)] for j in range(0,n)])
mat2 = np.array([[3 for i in range(0,m)] for j in range(0,k)])
start = timeit.default_timer()
mat = mat1.dot(mat2)
# print(mat)
end = timeit.default_timer()
print("Total time taken using numpy : ",end-start)
```

###### Here we can observe that matrix multiplication using numpy array is faster than using list of lists. Internally, list contains a pointer to list of pointers and these pointers point to pyObject that contains data item. While in case of numpy array, the matrix is stored as 2-D array and data items are stored at contiguous memory locations. Accessing the data items present at contiguous memory locations is much more efficient for storing and manipulating data. That's why numpy is faster than list of lists.

### Question 4
##### What is the gradient of the following function with respect to x and y?
###### Ans- 
$\frac{\partial f}{\partial x} = 2xy + y^3cos(x)$ ,    
$\frac{\partial f}{\partial y} = x^2 + 3y^2sin(x)$ 

hence gradient is 
$$\nabla f(x, y) = (\frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}) =  \left( 2xy + y^3 \cos(x), \, x^2 + 3y^2 \sin(x) \right)$$


### Question 5
##### Use sympy to confirm that you obtain the same gradient analytically.

```python
# %pip install --upgrade pip
# %pip install sympy 
import sympy as sp

x, y = sp.symbols('x y')

f = (x**2)*y + (y**3)*sp.sin(x)

dfdx = sp.diff(f, x)
dfdy = sp.diff(f, y)

grad = (dfdx, dfdy)
print("gradient is ", grad)
```

### Question 8
##### Store the same information using Python classes. We have an overall database which is a list of year objects. Each year contains a list of branches. Each branch contains a list of students. Each student has some properties like name, roll number and has marks in some subjects.

```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name 
        self.roll_no = roll_no 
        self.marks = marks

class Branch:
    def __init__(self, students):
        self.students = students 

class Year: 
    def __init__(self, branches):
        self.branches = branches 


student1 = Student('Shivansh Gupta', 24250086, {'Hindi':100, 'English':90, 'Math':100})
student2 = Student('Sreejan Gupta', 24250093, {'Hindi':10, 'English':9, 'Math':10})
branch = Branch([student1, student2])
year_2022 = Year([branch])
database = [year_2022]
print(database[0].branches[0].students[0].name)
        
```

### Question 10 
##### Using numpy generate a matrix of size 20X5 containing random numbers drawn uniformly from the range of 1 to 2. Using Pandas create a dataframe out of this matrix. Name the columns of the dataframe as a, b, c, d, e. Find the column with the highest standard deviation. Find the row with the lowest mean.

```python
import numpy as np
import pandas as pd
m, n = 20, 5

arr = np.random.uniform(1.0,2.0,m*n)
# print(arr)
mat = arr.reshape(m,n)
# print(mat)
# print(type(mat))

col_names = ['a', 'b', 'c', 'd', 'e']
df = pd.DataFrame(data = mat, columns = col_names)
# print(df)
print("Standard deviations for different columns are following :")
print(df.std(axis=0,ddof=0))
print("\nMean of different rows are : ")
print(df.mean(axis=1))

print("\nHere we can observe that column 'd' has highest standard deviation and row '0' has lowest mean.")
```

### Question 13 
##### Write a function to compute the argmin of a numpy array. The function should take a numpy array as input and return the index of the minimum element. You can use the np.argmin function to verify your solution.

```python
import numpy as np
def argmin(arr):
    sz = arr.size
    min_ind = 0
    min_ele = arr[0]
    for i in range(0,sz):
        if arr[i] < min_ele:
            min_ele = arr[i]
            min_ind = i

    return min_ind 

arr = np.array([23,67,12,56,8,90])
print(argmin(arr))
print(np.argmin(arr))

```
